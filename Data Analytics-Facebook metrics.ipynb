**Data Preprocessing**

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/DSBDA/Dataset/facebook.csv', sep=';')
```

```python
print(df.isnull().sum())
```

```python
df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
print(df.isnull().sum())
```

 Data Subsets

 Subset: Post Performance Based on Post Type = help to analyze which post Type gets more interactions.

```python
subset_post_performance = df[['Type', 'like', 'share', 'comment', 'Total Interactions']]
```

```python
subset_post_performance.head()
```

Subset: Engagement vs Time  = help  to find best times to post for higher engagement.

```python
subset_time_engagement = df[['Post Month', 'Post Weekday', 'Post Hour', 'Total Interactions']]
```

```python
subset_time_engagement.head()
```

```python
best_times_sorted = subset_time_engagement.sort_values(by='Total Interactions', ascending=False)
print(best_times_sorted.head(5))
```

subset:Post hour = help to see how many post are posted at particular time

```python
post_hour_counts = df['Post Hour'].value_counts().sort_index()
post_hour_counts.head()
```

  Merge Data  = merge subset Post Performance and Engagement vs Time

---



```python
merged = pd.merge(subset_time_engagement, subset_post_performance, on='Total Interactions')
merged.head()
```

Sort Data

```python
sorted_df_multi = df.sort_values(by=['Post Month', 'Post Hour'])
sorted_df_multi.head()
```

Transposing Data

```python
transposed_df = df.transpose()
transposed_df.head()
```

Shape and reshape Data

```python
print("Original Shape:", df.shape)
```

```python
reshaped_df = pd.melt(df, id_vars=['Type', 'Category'], value_vars=['like', 'comment', 'share'],var_name='Interaction Type', value_name='Count')
print(reshaped_df.shape)
reshaped_df.head()
```

```python
pivoted_df = reshaped_df.pivot_table(index=['Type', 'Category'],columns='Interaction Type',values='Count',aggfunc='sum').reset_index()
print(pivoted_df.shape)
pivoted_df.head()
```
